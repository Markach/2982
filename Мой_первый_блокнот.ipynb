## Корреляция

**Корреля́ция, или корреляцио́нная зави́симость** — статистическая взаимосвязь двух или более случайных величин, при этом изменения значений одной или нескольких из этих величин сопутствуют систематическому изменению значений другой или других величин.

Проще можно сказать, что ***корреляция*** - это мера того, насколько сильно две случайные величины связаны друг с другом. Значения колеблятся от **-1 до +1.**

**Значение 0** означает отсутствие корреляции. Значение необходимо интерпретировать, поскольку часто значение ниже -0,5 или выше 0,5 указывает на заметную корреляцию, а значения ниже этих значений предполагают менее заметную корреляцию.



Есть ли связь между стоимостью проезда и возрастом пассажира?

```python
df.Fare.corr(df.Age)
```

Результат 0,112328 показывает, что существует некоторая положительная корреляция. 

***Попробую провести более полный анализ:***

```python
# чтобы провести анализ, присвою числовое значение полу, мужской как 1 женский как 0
df["gender"] = 0
df.loc[df.Sex=='male', "gender"]=1
df.corr()
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
# df['f'].plot(kind="heatmap")
```

```python
df["gender"] = 0
df.loc[df.Sex=='male', "gender"]=1
corr = df.corr()
plt.figure(figsize=(11,8))
sns.heatmap(corr, cmap="Greens",annot=True)
plt.show()
```

```python
# Можно удалить избыточную повторяющуюся информацию
import numpy as np
df["gender"] = 0
df.loc[df.Sex=='male', "gender"]=1
corr.style.background_gradient(cmap='coolwarm', axis=None)
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True
corr[mask] = np.nan
(corr
 .style
 .background_gradient(cmap='coolwarm', axis=None, vmin=-1, vmax=1)
 .highlight_null(null_color='#f1f1f1')  # Color NaNs grey
 .set_precision(2))
```

Глядя выше на аннотированную тепловую карту (по умолчанию аннотация для коэффициента корреляции расположена в центре каждой ячейки)(даже не одна, а три для наглядности), можно сделать такие предварительные **выводы:**

1. положительная корреляция между стоимостью проезда и количеством родителей/детей на борту
2. положительная корреляция между стоимостью проезда и братьями и сестрами / супругами на борту
3. положительная корреляция между возрастом и тарифом на проезд

А также некоторая связь с полом и выжившим, но тут трудно это утверждать, потому что пол является бинарной переменной.

### Можно еще для наглядности попробовать поискать взаимосвязи другими способами


```python
# от количества членов семьи (супругов+ родственников) зависит выживаемость
sns.catplot('Siblings/Spouses Aboard', data=df, hue='Survived', kind='count')
```

На графике можно увидень несколько больший шанс выжить имели пассажиры у которых были родственники на корабле, напр. супруг. 

```python
# от количества членов семьи (детей+ родителей) зависит выживаемость
sns.catplot('Parents/Children Aboard', data=df, hue='Survived', kind='count')
```

Аналогичная взаимосвязь если смотреть связку дети-родители.

```python
# следовательно можно проследить, что мужчины умирали чаще
sns.catplot('Sex', data=df, hue='Survived', kind='count')
```

```python
# можно проследить еще связь с гендером и классом кают в которых ехели пассажиры
sns.catplot(x='Sex', y='Survived', col='Pclass', data=df, kind='bar')
```

Из графика выше можно увидеть, что бедных выживших мужчин было меньше всего.

```python
# еще одно распределение зависимости выживаемости от цены билета
sns.catplot('Fare', data=df, hue='Survived', kind='count')
```
